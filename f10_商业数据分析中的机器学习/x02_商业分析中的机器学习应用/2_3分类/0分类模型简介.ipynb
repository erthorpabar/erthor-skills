```python
''' 
————————混淆矩阵————————
          预测为a     预测为b
标签为a   对a (TP)    错a→b (FN)
标签为b   错b→a (FP)  对b (TN) 


           预测类别
         |  a  |  b  |
实际  a  | TP  | FN  |  
类别  b  | FP  | TN  |



有多少输入被正确分类,越接近1越好
准确率 = (TP + TN) / (TP + TN + FP + FN)

预测为正的输入,有多少是正确的,越接近1越好
精确率 = TP / (TP + FP)

实际为正的输入,有多少被正确分类,越接近1越好
召回率（查全率） = TP / (TP + FN)
当一个正样本被漏掉(a->b) 的代价很高时候 需要额外注意这个指标

预测为负的输入,有多少被正确分类,越接近1越好
特异度 = TN / (TN + FP)

F1 = 2 * 精确率 * 召回率 / (精确率 + 召回率)
越接近1越好

————————roc/auc 曲线————————

tp + fn = 1
fp + tn = 1

roc曲线
纵坐标 = tp(rate) =  tp/(tp+fn)
横坐标 = fp(rate) =  fp/(fp+tn)
越向上凸 越好
0,0 - 1,1 对角线,代表持平,代表没有分类
越向下凹 越不好

auc = 曲线下面积
auc越高 模型越好
0.9-1.0 极好
0.8-0.9 优秀
0.7-0.8 一般
0.6-0.7 较差

通过调整分类的阈值 可以调整roc曲线


'''
```

```python
''' 
信息熵
0 基础概念
m = 可能的选项(信息量)
p = 选项的概率(概率)
比如
p(a1) = 1/2
p(a2) = 1/2
m=2代表有两个选项
mp = 1
m*p(a1) + m*p(a2) = 1

1 抛一次硬币
p(a1) = 1/2
p(a2) = 1/2
最高效率 每次排除一半信息量 需要1次确认结果
整体信息熵n = log2(m=2) = 1 bit
m = 1/p
对于单个信息熵
n(a1) = p * log2(1/p) = 1/2 * log2(2) = 0.5 bit
n(a2) = p * log2(1/p) = 1/2 * log2(2) = 0.5 bit
整体信息熵 n = n(a1) + n(a2) = 1 bit

2 抛两次硬币
p(a1) = 1/4
p(a2) = 1/4
p(a3) = 1/4
p(a4) = 1/4
最高效率 每次排除一半信息量 需要2次确认结果
整体信息熵n = log2(m=4) = 2 bit
m = 1/p
对于单个信息熵
n(a1) = p * log2(1/p) = 1/4 * log2(4) = 0.5 bit
n(a2) = p * log2(1/p) = 1/4 * log2(4) = 0.5 bit
n(a3) = p * log2(1/p) = 1/4 * log2(4) = 0.5 bit
n(a4) = p * log2(1/p) = 1/4 * log2(4) = 0.5 bit
整体信息熵 n = n(a1) + n(a2) + n(a3) + n(a4) = 2 bit

+++++++++++++++++++++
总结出公式
n = sum(p * log2(1/p))
其中
m = 1/p
+++++++++++++++++++++

3 对于非等概率的情况
p(a1) = 4/8
p(a2) = 2/8
p(a3) = 1/8
p(a4) = 1/8
不能通过m之间计算整体信息熵n
对于单个信息熵
n(a1) = p * log2(1/p) = 4/8 * log2(8/4) = 1/2 bit
n(a2) = p * log2(1/p) = 2/8 * log2(8/2) = 1/4 bit
n(a3) = p * log2(1/p) = 1/8 * log2(8/1) = 3/8 bit
n(a4) = p * log2(1/p) = 1/8 * log2(8/1) = 3/8 bit
整体信息熵 n = n(a1) + n(a2) + n(a3) + n(a4) = 1.75 bit

4 此时n = 1.75 bit
当所有概率相等时候 代表无区分度 则熵最大
等概率的最高熵
n(max) = log2(m)



5 另一种计算信息熵的方式
p(a1) = 1/2
p(a2) = 1/4
p(a3) = 1/8
p(a4) = 1/8
想象一个决策树
   / \
a1   / \
   a2  / \
      a3  a4

当第n= log2(m=2) = 1 次决策时 分类出a1
当第n= log2(m=4) = 2 次决策时 分类出a2
当第n= log2(m=8) = 3 次决策时 分类出a3,a4

单个信息熵 = n/m 
n = 总决策次数
m = 排除的选项数量
n/m = 平均每个选项所需的决策次数
信息熵 = 平均识别一个事件所需的最小判断次数
信息熵 ≈ 决策树中的平均路径长度

用m计算n
n1 = 1
n2 = 2
n3 = 3
n4 = 3
m1 = 2
m2 = 2^2 = 4
m3 = 2^3 = 8
m4 = 2^3 = 8
= sum(n/m) 
= n1/m1 + n2/m2 + n3/m3 + n4/m4
= 1/2 + 2/4 + 3/8 + 3/8
= 1.75

用p计算n
= sum(pn)
= p1 * n1 + p2 * n2 + p3 * n3 + p4 * n4
= 1/2 * 1 + 1/4 * 2 + 1/8 * 3 + 1/8 * 3
= 1.75

6 交叉熵
正确分类的熵
 / \
A  / \
   B  C
p(A) = 0.5
p(B) = 0.25
p(C) = 0.25
n(A) = 1
n(B) = 2
n(C) = 2
信息熵n
= p(A) * n(A) + p(B) * n(B) + p(C) * n(C)
= 0.5 * 1 + 0.25 * 2 + 0.25 * 2
= 1.5


错误分类的熵(弄错了决策路径,但是概率是不变的,导致平均决策次数增加)
 / \
C  / \
   B  A
p(A错) = 0.25
p(B错) = 0.25
p(C错) = 0.5
n(A错) = 2
n(B错) = 2
n(C错) = 1

交叉熵h
= p(A) * n(A错) + p(B) * n(B错) + p(C) * n(C错)
= 0.5 * 2 + 0.25 * 2 + 0.25 * 1
= 1.75

n = log2(1/p)
使用p计算交叉熵h
= p(A) * log2(1/p(A错)) + p(B) * log2(1/p(B错)) + p(C) * log2(1/p(C错))
= 0.5 * log2(1/0.25) + 0.25 * log2(1/0.25) + 0.25 * log2(1/0.5)
= 0.5 * 2 + 0.25 * 2 + 0.25 * 1
= 1.75

7 相对熵
相对熵 = 交叉熵 - 信息熵


8 总结
熵 = 决策树的平均路径长度, 即平均决策次数(这代表好的决策树 应优先考虑最具区分度的判断条件)

m为概率的最大公约数
m不变下
1 最大熵 = 1 无任何决策依据 2 各选项无任何区分度 3 平均决策次数最大 4 熵最大
2 最小熵(信息熵) = 1 最优决策依据 2 各选项有最佳区分度 3 平均决策次数最小 4 熵最小
3 实际熵(交叉熵) = 1 非最优决策依据 2 各选项非最佳区分度 3 平均决策次数介于最大最小之间 4 熵介于最大最小之间
4 相对熵 = 交叉熵 - 信息熵

举例 
最大熵
  / \
/ \  / \
A  A B  C
n(max) 
= p(A)*n(A) +p(A)*n(A) + p(B)*n(B) + p(C)*n(C)
= 0.25*2 + 0.25*2 + 0.25*2 + 0.25*2
= 2

最小熵
 / \
A  / \
   B  C
n(min) 
= p(A)*n(A) + p(B)*n(B) + p(C)*n(C)
= 0.5*1 + 0.25*2 + 0.25*2
= 1.5

交叉熵
 / \
C  / \
   B  A
n(cross)
= p(A)*n(A错) + p(B)*n(B错) + p(C)*n(C错)
= 0.25*2 + 0.25*2 + 0.5*1
= 1.75

相对熵
= n(cross) - n(min)
= 1.75 - 1.5
= 0.25

'''
```
