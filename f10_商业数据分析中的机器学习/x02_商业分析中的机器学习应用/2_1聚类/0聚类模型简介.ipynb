```python
'''核心概念
1
样本之间的距离
1 欧几里得距离 = c = √(a^2+b^2)
2 曼哈顿距离 c = a + b
3 完全平方距离 c^2 = a^2 + b^2
4 weight c = √(w1*a^2 + w2*b^2)

2
聚类之间的距离
单链接 = 样本中最近的两点
全链接 = 样本中最远的两点
平均链接 = 质心的距离
ward方法 = 质心的距离 * √(2AB/(A+B))

3 找出最佳的k值
0 霍普金斯指数，确定数据集有聚类倾向的指标
1 经验值 k = √(n/2)
2 手肘法
遍历所有k值 并聚类
计算 x - 质心 的距离平方和
k越多 距离平方和越小
当增加k 距离平方和不再显著下降
代表 增加k 对 分类合理性的贡献不再增加
此时k为最佳分类数量
即 拐点 = 最佳k
3 轮廓系数
遍历所有k值 并聚类
a = 样本 到 当前聚类所有点的平均距离
b = 样本 到 最近聚类所有点的平均距离
轮廓系数 s = (b-a)/max(a,b)
s 越接近1 聚类效果越好
s 越接近-1 聚类效果越差
s 越接近0 聚类效果一般


4 评估聚类效果
1 紧凑性 = 样本 - 质心 距离的平方和
2 分离性 = 质心 - 质心 距离
方法
1 调整兰德指数ARI
2 归一化互信息NUI
3 纯度
'''
```
