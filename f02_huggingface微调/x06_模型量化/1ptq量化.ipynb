0 导入包

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, GPTQConfig
```

1 ptq量化

```python
model_id = "facebook/opt-125m"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    # 模型路径
    model_id, 

    # 参数与显存
    device_map={"": 0}, # k代表参数名前缀，空代表所有模型参数 v代表在哪张gpu上运行

    # 量化加载配置
    quantization_config = GPTQConfig(
        bits=4, # 量化精度
        group_size = 128, # 推荐128，一组有128给token
        dataset = "wikitext2", # 用来校准的数据集，可填写这些参数'wikitext2','c4','c4-new','ptb','ptb-new'
        # dataset =  ["auto-gptq is an easy-to-use model quantization library with user-friendly apis, based on GPTQ algorithm."] # 自己造矫正数据集，当然这个案例效果不会好
        tokenizer = tokenizer,
        ),
)
```

2 保存

```python

model.save_pretrained("opt-125m-gptq") 
tokenizer.save_pretrained("opt-125m-gptq") 
```
