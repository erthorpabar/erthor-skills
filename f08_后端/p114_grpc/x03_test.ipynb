```python
# 定义异步请求函数
import grpc
from x01_pb2 import in_message, out_message
from x01_pb2_grpc import xyzStub 
'''一个stub代表一个不同的服务 可引入多个stub代表同意端口下不同路由服务'''

async def ask(input_value):
    async with grpc.aio.insecure_channel('localhost:50051') as channel:
        stub = xyzStub(channel=channel)
        request = in_message()
        request.x = input_value  # 赋值输入
        response = await stub.Search(request)
        return response
```

```python
import time 
import asyncio
```

```python
# 单次请求
start = time.time()

res = await ask("hello")

end = time.time()

print(f"{end - start:.2f}秒")
print(res.var)
```

```python
start = time.time()

# 并发执行两个请求
tasks = [ask("hello"), ask("world")]
results = await asyncio.gather(*tasks)

res1, res2 = results

end = time.time()
print(f"{end - start:.2f}秒")

print(res1.var)
print(res2.var)
```
