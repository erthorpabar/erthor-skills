```python
import requests
import time  # 导入 time 模块

start_time = time.time() # 开始时间

# # 测试请求数据
url = "http://127.0.0.1:7005/two_character/gpu/01"  # 根据你的服务地址调整

prompt = '''
a girl
black beret,
Long golden hair,
Black turtleneck sweater,
white pants,
'''

data = {
    "prompt": prompt,
    "width": 1024,
    "height": 1024,
    "checkpoint_name": "flux-dev-fp8.safetensors",
    "lora_name": "flux_lora/推文韩漫都市言情_V1.0.safetensors",
}

# 发送 POST 请求
response = requests.post(url, json=data)


end_time = time.time() # 结束时间



total_time = end_time - start_time # 总耗时
print(f"生图+返图耗时: {total_time:.2f} 秒")

print(response.json())
```
