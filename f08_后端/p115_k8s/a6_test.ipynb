```python
# —————————————————————get接口测试————————————————————————
import requests
import time  # 导入 time 模块

start = time.time() # 开始时间


# # 测试请求数据
url = "http://127.0.0.1:80/aaa"  # 根据你的服务地址调整

# 发送 POST 请求
response = requests.get(url)


end = time.time() # 结束时间
print(f"耗时: {end-start:.2f} 秒")

print(response.json())
```
