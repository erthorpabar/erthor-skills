```python
import requests
import json
base_url = "http://127.0.0.1:7004"
```

```python
# 1. 测试根路径 GET 请求
response = requests.get(f"{base_url}/")
print(response.json())
```

```python
# 2. 测试根路径 POST 请求
response = requests.post(f"{base_url}/")
print(response.json())
```

```python
# 3. 测试组路由 GET /v1/
response = requests.get(f"{base_url}/v1/")
print(response.json())
```

```python
# 4. 测试用户数据验证 POST /user
user_data = {
    "name": "张三",
    "age": 25
}
response = requests.post(f"{base_url}/user", json=user_data)
print(response.json())
```

```python
# 5. 测试创建用户 POST /sql/create
create_data = {
    "name": "测试用户",
    "password": "123456"
}
response = requests.post(f"{base_url}/sql/create", json=create_data)
print(response.json())
```

```python
# 6. 测试查询所有用户 POST /sql/find
response = requests.post(f"{base_url}/sql/find")
from pprint import pprint
pprint(response.json())
```

```python
# 7. 测试更新用户 POST /sql/update
update_data = {
    "id": 1,  # 假设要更新ID为1的用户
    "name": "更新后的用户名",
    "password": "新密码123"
}
response = requests.post(f"{base_url}/sql/update", json=update_data)
print(response.json())
```

```python
# 8. 测试删除用户 POST /sql/delete
delete_data = {
    "id": 1  # 假设要删除ID为1的用户
}
response = requests.post(f"{base_url}/sql/delete", json=delete_data)
print(response.json())
```

```python
# 9. 测试错误情况 - 缺少必填字段
invalid_data = {
    "name": "只有姓名"
    # 缺少 age 字段
}
response = requests.post(f"{base_url}/user", json=invalid_data)
print(f"状态码: {response.status_code}")
print(response.json())
```
